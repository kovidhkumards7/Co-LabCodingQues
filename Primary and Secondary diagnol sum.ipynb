```
n = int(input("enter the size of square matrix  "))
a = [[int(input("enter element "))for i in range(n)]for j in range(n)]
print("the entered matrix is  ")
for i in range(n):
  for j in range(n):
    print(a[i][j],end="  ")
  print()
pd,sd = 0,0
for i in range(n):
  for j in range(n):
    if (i==j):
      pd += a[i][j]
    if (i+j==n-1):
      sd += a[i][j]
print("primary diagnol",pd)
print("secondary diagnol",sd)
```
