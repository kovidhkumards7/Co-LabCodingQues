Check whether the letter "p" comes three times consecutively in a given string


**input**: pineapple                  
**output**: false (p didnt occur three times simultaneously)                 

**input**: anppp             
**output**: true          

```python
#your code goes here
```

```python
#@title
word="anppp"
result=0
for i in range(len(word)-2):
  if word[i]=='p' and word[i+1]=='p' and word[i+2]=='p':
    result=1
if result==1:
  print("true")
else:
  print("false")
```
