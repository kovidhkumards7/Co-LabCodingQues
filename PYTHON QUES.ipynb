```
#ques 11
import math
arr = input("enter comma seperated series of numbers  ")
binary = []
binary = arr.split(",")
for i in binary:
  c=i[::-1]
  s=0
  for j in range(0,len(c)):
    s=s+(int(c[j])*math.pow(2,j))
  if (s%5==0):
    print(i,end=",")
```

```
#ques 18
a = input("enter password  ")
u,l,n,s=0,0,0,0
for i in a:
  if (i.isupper()):
    u+=1
  elif (i.islower()):
    l+=1
  elif (i.isdigit()):
    n+=1
  else:
    s+=1
if (u>=1) and (l>=1) and (n>=1) and (s>=1) and (len(a)>6) and (len(a)<12):
  print("password accepted")
else:
  print("password not accepted")
```
